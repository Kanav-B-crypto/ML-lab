```python
import numpy as np

# (a) Reverse the NumPy array
arr = np.array([1, 2, 3, 6, 4, 5])
print(arr[::-1])

# (b) Flatten the NumPy array
array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
print(array1.flatten())
print(array1.ravel())

# (c) Compare two NumPy arrays
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
print(np.array_equal(arr1, arr2))
print(arr1 == arr2)

# (d) Most frequent value and indices
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])
print(np.bincount(x).argmax(), np.where(x == np.bincount(x).argmax()))
print(np.bincount(y).argmax(), np.where(y == np.bincount(y).argmax()))

# (e) Matrix sums
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(np.sum(gfg))
print(np.sum(gfg, axis=1))
print(np.sum(gfg, axis=0))

# (f) Matrix properties
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
print(np.trace(n_array))
vals, vecs = np.linalg.eig(n_array)
print(vals)
print(vecs)
print(np.linalg.inv(n_array))
print(np.linalg.det(n_array))

# (g) Matrix multiplication and covariance
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])
print(np.matmul(p, q))
print(np.cov(p, q))

p = np.array([[1, 2], [2, 3], [4, 5]])
q = np.array([[4, 5, 1], [6, 7, 2]])
print(np.matmul(p, q))
print(np.cov(p.T, q))

# (h) Inner, outer, cartesian product
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
print(np.inner(x, y))
print(np.outer(x, y))
print(np.kron(x, y))
```

```python
import numpy as np

# (a) For the array
array = np.array([[1, -2, 3],[-4, 5, -6]])

# (i) Element-wise absolute value
print(np.abs(array))

# (ii) Percentiles of flattened, column-wise, row-wise
print(np.percentile(array, [25, 50, 75]))          # flattened
print(np.percentile(array, [25, 50, 75], axis=0))  # per column
print(np.percentile(array, [25, 50, 75], axis=1))  # per row

# (iii) Mean, Median, Standard Deviation
print(np.mean(array), np.median(array), np.std(array))                # flattened
print(np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0))  # columns
print(np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1))  # rows

# (b) For array a
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Floor
print(np.floor(a))

# Ceiling
print(np.ceil(a))

# Truncated
print(np.trunc(a))

# Rounded
print(np.round(a))
```

```python
import numpy as np

# (a) For the array
array = np.array([10, 52, 62, 16, 16, 54, 453])

# (i) Sorted array
print(np.sort(array))

# (ii) Indices of sorted array
print(np.argsort(array))

# (iii) 4 smallest elements
print(np.partition(array, 3)[:4])

# (iv) 5 largest elements
print(np.partition(array, -5)[-5:])

# (b) For the array
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# (i) Integer elements only
print(array[array == array.astype(int)])

# (ii) Float elements only
print(array[array != array.astype(int)])
```
